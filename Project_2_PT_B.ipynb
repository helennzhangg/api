```python
#We will be using pandas for loading the csv file into a dataframe
import pandas as pd
#matplotlib is a python library for basic graph plotting
import matplotlib.pyplot as plt

#seaborn is another python graphing library that is built upon matplotlib and allows you to optimize matplotlib's output
import seaborn as sns 

#switching from matplotlib's default graphing style to that of seaborn's. Seaborn displays graph lines, matplotlib doesn't
sns.set(color_codes = True)

#removing warnings from the output
import warnings
warnings.filterwarnings("ignore")
```

```python
#link to download csv file: https://www.kaggle.com/datasets/muhmores/spotify-top-100-songs-of-20152019?select=Spotify+2010+-+2019+Top+100.csv
#constently upload + copy path from the csv file when closing google chrome
#loading csv data
music = pd.read_csv('/content/Spotify 2010 - 2019 Top 100.csv')
```

```python
music
```

```python
#limiting the amount of rows to the first 50 to give a better show of the graphs 
limitedmusic = music.head(50)
limitedmusic
```

```python
#showing that there are more dance pop genres in the top 50
limitedmusic['top genre'].value_counts()
```

```python
#comparing beats per minute to the energy 
limitedmusic.plot(kind = 'scatter', x = 'bpm', y= 'nrgy')
```

```python
#comparing beats per minute to danceability
limitedmusic.plot(kind = 'scatter', x = 'bpm', y= 'dnce')
```

```python
#comparing beats per minute to decibels 
limitedmusic.plot(kind = 'scatter', x = 'bpm', y= 'dB')
```

```python
#comparing energy to danceability in terms of the genre 
sns.FacetGrid(limitedmusic, hue='top genre', palette = 'Set2', size = 10).map(plt.scatter, 'dnce', 'nrgy').add_legend()
```

```python
#comparing energy to danceability in terms of the artist
sns.FacetGrid(limitedmusic, hue='artist', palette = 'Set2', size = 10).map(plt.scatter, 'nrgy', 'live').add_legend()
```

```python
#x axis rotation code reference: https://stackoverflow.com/questions/26540035/rotate-label-text-in-seaborn-factorplot
```

```python
#comparing the genre to danceability 
sns.boxplot(x='top genre', y= 'dnce', palette = 'Set2', data = limitedmusic)
plt.xticks(rotation=90)
```

```python
#comparing the genre to energy
sns.boxplot(x='top genre', y= 'nrgy', palette = 'Set2', data = limitedmusic)
plt.xticks(rotation=90)
```

```python
#comparing the genre to beats per minute
sns.boxplot(x='top genre', y= 'bpm', palette = 'Set2', data = limitedmusic)
plt.xticks(rotation=90)
```

```python
#comparing the genre to decibels
sns.boxplot(x='top genre', y= 'dB', palette = 'Set2', data = limitedmusic)
plt.xticks(rotation=90)
```

```python
#comparing the genre to liveliness 
sns.boxplot(x='top genre', y= 'live', palette = 'Set2', data = limitedmusic)
plt.xticks(rotation=90)
```

```python
#density of beats per minute in terms of genres 
sns.FacetGrid(limitedmusic, hue='top genre', palette = 'Set2', size=10).map(sns.kdeplot, 'bpm').add_legend()
```

```python
#density of liveliness in terms of artists 
sns.FacetGrid(limitedmusic, hue='artist', palette = 'Set2', size=10).map(sns.kdeplot, 'live').add_legend()
```
