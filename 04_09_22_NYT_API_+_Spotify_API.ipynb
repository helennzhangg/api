```python
import requests 
import json 
```

```python
key = 'zsFAqhZIGshKveJc1qdp4fq1GVFe8DFp'
#aoi end point
#parameter : stephen king
url = 'https://api.nytimes.com/svc/books/v3/reviews.json?author=Stephen+King&api-key=' + key

r = requests.get(url)
#r.status_code ; status code = 200
if (r.status_code == 200):
  data = r.json()
  formatted_json = json.dumps(data, sort_keys = True, indent = 10)
  print(formatted_json)
```

```python
#pandas dataframe with book title, date, url
#export dataframe as csv file
book_name = []
pub_date = []
review_url = []
```

```python
#create a for loop because there is a list inside the dictionary 
  #you cannot do data['results']['book_title'] because there will be an error
for book in data['results']:
  print(book)
  book_name.append(book['book_title'])
  pub_date.append(book['publication_dt'])
  review_url.append(book['url'])
```

```python
import pandas as pd
```

```python
titledata = {
    'Book Title' : book_name,
    'Publication Date' : pub_date,
    'Url Review' : review_url
}
print(titledata)
```

```python
books = pd.DataFrame(titledata)
books
```

```python
#drop index : index = False; we drop the index because the index is not helpful to readers 
books.to_csv('name_date_url.csv', index = False)
```
